1.Give a real-world scenario from your dataset above that can be modeled using the binomial distribution.

```python
import pandas as pd
from scipy.stats import binom

# Define success criteria (revenue > $300 million)
revenue_threshold = 300 # here in our dataset the units is  millions

# Calculate the count of successful movies (movies with revenue > $300 million)
successful_movies = imdb[imdb['Revenue (Millions)'] > revenue_threshold]
count_successful_movies = len(successful_movies)

#Total number of movies
total_movies = len(imdb)

#Calculate the probability of success (proportion of movies with revenue > $300 million)
p = count_successful_movies / total_movies

# Number of trials (total number of movies)
n = total_movies

# Create a binomial distribution
binomial_dist = binom(n, p)

#Now we want to calculate the probability of exactly k successes (e.g., exactly 50 successful movies):
k = 50  
probability_exact_k_successes = binomial_dist.pmf(k)

# what's the  probablity to get more than 50 successful movie 
probability_more_than_k_successes = 1 - binomial_dist.cdf(k)

# Output results
print(f"Probability of exactly {k} successful movies: {probability_exact_k_successes:.4f}")
print(f"Probability of more than {k} successful movies: {probability_more_than_k_successes:.4f}")

```

1. Normal distribution Analysis

```python
runtime_mean= imdb['Runtime (Minutes)'].mean()
runtime_sd = imdb['Runtime (Minutes)'].std()
print(runtime_mean)
print(runtime_sd)
```

```python
#importing libraries and packages
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = (8, 6) #setting the size of the distribution

# plot the histogram with density normalized
imdb_hist = sns.distplot(imdb['Runtime (Minutes)'], norm_hist = True, kde = False, color = 'purple', bins = 7)

# calculate the x value and y value for the normal distribution 
x = np.arange(0, 200)                                                  
y = norm.pdf(x, loc =runtime_mean , scale = runtime_sd) # setting the mean as loc and the Standard deviation as scale.

# plotting the normal distribution curve 
imdb_hist.plot(x, y, 'red');

#Add the title and label for normal distribution
plt.title("Normal Distribution for Runtime",fontsize=15)
plt.xlabel("Runtime (Minutes)",fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.show()
```

Now according the normal distribution, we wand to calculate the probablity less than 100(P< 100) and the probablity between 100 and 125(100< P < 150) and the probability greater than 150 (P > 150). So we can calcualte the CDF(cumulative distribution function)

```python
# calculate the probablity of runtime less than 100 (P <100)

# Set the size of the plot
plt.rcParams['figure.figsize'] = (8, 6)

# Create the range of x values 
x = np.arange(0,200)

# Calculate the corresponding y values for the normal distribution
y = norm.pdf(x, loc=runtime_mean, scale= runtime_sd)

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution of runtime', color='blue')

# Shade the area under the curve for value less than 100
x_shaded = np.linspace(0, 100, 500) 
y_shaded = norm.pdf(x_shaded, loc=runtime_mean, scale=runtime_sd)
plt.fill_between(x_shaded, y_shaded, color='purple', alpha=0.5, label='Shaded Area :P(x < 100)')

# Add title and labels
plt.title("Normal Distribution of runtime", fontsize=15)
plt.xlabel("runtime", fontsize=12)
plt.ylabel("Density", fontsize=12)

# Add a legend
plt.legend()
plt.ylim(bottom=0)

# Show the plot

plt.show()

# calculate the probablity of runtime less than 100(P <100)
p_less_than_100 = norm.cdf(100, loc=runtime_mean, scale=runtime_sd)
print(f"P(x < {100}) = {p_less_than_100}")
```

```python
# calculate the probablity of runtime between  100 and 150 minutes P:(100<X<150)
# Set the size of the plot
plt.rcParams['figure.figsize'] = (8, 6)

# Create the range of x values 
x = np.arange(0, 200)

# Calculate the corresponding y values for the normal distribution
y = norm.pdf(x, loc=runtime_mean, scale= runtime_sd)

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution of runtime', color='blue')

# Shade the area under the curve between 100 and 150
x_shaded = np.linspace(100, 150, 500) 
y_shaded = norm.pdf(x_shaded, loc=runtime_mean, scale=runtime_sd)
plt.fill_between(x_shaded, y_shaded, color='purple', alpha=0.5, label='Shaded Area :P(100< x < 150)')

# Add title and labels
plt.title("Normal Distribution of runtime ", fontsize=15)
plt.xlabel("Runtime(minutes)", fontsize=12)
plt.ylabel("Density", fontsize=12)

# Add a legend
plt.legend()
plt.ylim(bottom=0)

plt.show()


# calculate the probablity of runtime between 100 and 150 P(100<x<150)
p_between_100_150= norm.cdf(150, loc=runtime_mean, scale=runtime_sd)-norm.cdf(100,loc=runtime_mean,scale=runtime_sd)
print(f"P({100}< x < {150}) = {p_between_100_150:.4f}")
```

```python
# calculate the probablity of value greater  than 150 (P >150)
# Set the size of the plot
plt.rcParams['figure.figsize'] = (8, 6)

# Create the range of x values 
x = np.arange(0, 200)

# Calculate the corresponding y values for the normal distribution
y = norm.pdf(x, loc=runtime_mean, scale= runtime_sd)

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution of runtime', color='blue')

# Shade the area under the curve greater than 150
x_shaded = np.linspace(150,200,500) 
y_shaded = norm.pdf(x_shaded, loc=runtime_mean, scale=runtime_sd)
plt.fill_between(x_shaded, y_shaded, color='purple', alpha=0.5, label='Shaded Area :P(x>150)')

# Add title and labels
plt.title("Normal Distribution of runtime ", fontsize=15)
plt.xlabel("Runtime(minutes)", fontsize=12)
plt.ylabel("Density", fontsize=12)

# Add a legend
plt.legend()
plt.ylim(bottom=0)

plt.show()

# calculate the probablity of runtime greater than 150 P(X>150)
p_greater_than_150=1- norm.cdf(150, loc=runtime_mean, scale=runtime_sd)
print(f"P(x> {150}) = {p_greater_than_150:.4f}")
```

2. Revenue (Millions) Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

#setting the size of the distribution
plt.figure(figsize=(10, 6))
revenue = imdb['Revenue (Millions)'].dropna() #dropna() is used to remove the missing values in the dataframe.

# calculating the mean and standard devision of the Revenue
mean = revenue.mean()
std = revenue.std()
print(mean)
print(std)

# Use Seaborn to get a normalized histogram
sns.histplot(revenue, bins=30, stat="density", color='green', edgecolor='blue', alpha=0.6)

# plot the normal distribution curve 
x = np.linspace(revenue.min(), revenue.max(), 100) #linspace() is used to generate the array of evenly spaced numbers for a particular range.Storing the minimum and maximum of revenue in x array.
y = norm.pdf(x, loc=mean, scale=std) #pdf is the probability density function
plt.plot(x, y, color='black', label='Normal Distribution')

# Adding labels and title
plt.title("Normal Distribution for Revenue", fontsize=15)
plt.xlabel("Revenue (Millions)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.show()
```

```python
#Take two values a and b,a is 80 and b is 160
a, b = 80, 160  

# calculate the probablity of a movie revenue less than 6 (P <80)
p_less_a = norm.cdf(a, mean, std)
print(f"P(X < {a}) = {p_less_a:.3f}")  # Rounds a value as a floating point number with 3 decimal points

# calculate the probablity of a movie revenue between 80 and 160 (80<P<160)
p_between_a_b = norm.cdf(b, mean, std) - norm.cdf(a, mean, std) #cdf is the cummulative density function and finding the differnece of the cdf of a and b
print(f"P({a} < X < {b}) = {p_between_a_b:.3f}")


# calculate the probablity of a movie revenue greater than 160(P>160)
p_greater_b = 1 - norm.cdf(b, mean, std) #to find the probabilty of greater than 160,we are subtracting 1 from the cdf of b. 
print(f"P(X > {b}) = {p_greater_b:.3f}")
```
